# Benchmark LLM — Agricultura Brasileira

Notebook interativo para avaliar modelos LLM via Ollama em Q&A agrícola, com tracking via MLflow.

Toda a lógica está em `src/` — este notebook apenas configura e executa.

```python
import os
import sys

# Ensure project root is on the path when running from notebooks/
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

os.environ["MLFLOW_ENABLE_SYSTEM_METRICS_LOGGING"] = "true"
```

```python
import mlflow
from src.config import MLFLOW_TRACKING_URI, PROMPTS
from src.data_loader import load_qa_dataset
from src.evaluation import run_evaluation

mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)
```

## Configuração

Altere as variáveis abaixo para mudar modelo, prompt ou dataset.

```python
MODEL = "llama3.1:8b"
PROMPT_VERSION = "v3"
DATASET = "citrus_sugarcane_qa.json"

prompt_template = PROMPTS[PROMPT_VERSION]
eval_data = load_qa_dataset(DATASET)

print(f"Model:   {MODEL}")
print(f"Prompt:  {PROMPT_VERSION}")
print(f"Dataset: {DATASET} ({len(eval_data)} questions)")
```

## Executar avaliação

```python
results = run_evaluation(
    eval_data,
    model=MODEL,
    prompt_template=prompt_template,
    prompt_version=PROMPT_VERSION,
)

print("Averages:")
for name, value in results["averages"].items():
    print(f"  {name}: {value:.4f}")
```
